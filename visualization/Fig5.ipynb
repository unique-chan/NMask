```python
import numpy as np
import matplotlib.pyplot as plt

# X axis (Kernel)
kernels = ['3x3', '5x5', '7x7']
x = np.arange(len(kernels))
width = 0.35

# Data (Vaihingen)
# Baseline
opening_base = [76.49, 75.89, 75.40]   # E->D (Opening)
opening_nmask = [76.83, 76.93, 76.47]  


# NMask
closing_base = [76.12, 75.99, 75.28]   # D->E (Closing)
closing_nmask = [76.87, 76.56, 76.43]  

def plot_one(y_base, y_nmask, xlabel, save_name, ylim=None):
    plt.figure(figsize=(4, 5))
    ax = plt.gca()

    ax.bar(x + width/2, y_base, width,
           label='Baseline',
           color='#1f77b4')

    ax.bar(x - width/2, y_nmask, width,
           label='NMask',
           color='#ff7f0e',
           hatch='//')

    ax.set_xlabel(xlabel, fontsize=13)
    ax.set_ylabel('Test mIoU (%)', fontsize=13)
    ax.set_xticks(x)
    ax.set_xticklabels(kernels)

    if ylim is not None:
        ax.set_ylim(*ylim)

    ax.grid(axis='y', linestyle='--', linewidth=1, alpha=0.7)
    ax.legend(fontsize=12)

    plt.tight_layout()
    plt.savefig(save_name, format="svg", bbox_inches="tight")
    plt.show()

# Opening plot
plot_one(opening_base, opening_nmask,
         xlabel='Kernel size',
         save_name='Fig5-3-opening-vaihingen.svg',
         ylim=(75, 77.5))

# Closing plot
plot_one(closing_base, closing_nmask,
         xlabel='Kernel size',
         save_name='Fig5-3-closing-vaihingen.svg',
         ylim=(75, 77.5))
```
